```python
import sys
import os

# Add project root to Python path
sys.path.append(os.path.abspath(".."))

import matplotlib.pyplot as plt

from src.evaluate import (
    evaluate_keras,
    evaluate_pytorch
)

from src.metrics import (
    print_metrics,
    plot_roc_curve
)

from src.config import CLASS_LABELS


def run_model_comparison():
    """
    Run full comparison between Keras and PyTorch models.
    """

    print("\n==============================")
    print("MODEL COMPARISON STARTED")
    print("==============================\n")

    try:
        # Evaluate Keras model
        print("Evaluating Keras model...\n")

        keras_labels, keras_preds, keras_probs = evaluate_keras()

        print("Keras evaluation complete.\n")

        # Evaluate PyTorch model
        print("Evaluating PyTorch model...\n")

        pytorch_labels, pytorch_preds, pytorch_probs = evaluate_pytorch()

        print("PyTorch evaluation complete.\n")

        print("==============================")
        print("METRICS REPORT")
        print("==============================\n")

        # Print Keras metrics
        print_metrics(
            keras_labels,
            keras_preds,
            keras_probs,
            CLASS_LABELS,
            "Keras Model"
        )

        # Print PyTorch metrics
        print_metrics(
            pytorch_labels,
            pytorch_preds,
            pytorch_probs,
            CLASS_LABELS,
            "PyTorch Model"
        )

        print("\n==============================")
        print("PLOTTING ROC CURVES")
        print("==============================\n")

        # ROC curve comparison
        plt.figure(figsize=(8, 6))

        plot_roc_curve(
            keras_labels,
            keras_probs,
            "Keras"
        )

        plot_roc_curve(
            pytorch_labels,
            pytorch_probs,
            "PyTorch"
        )

        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title("ROC Curve Comparison")
        plt.legend()
        plt.grid(True)
        plt.show()

        print("\nModel comparison completed successfully.")

    except FileNotFoundError as error:
        print(f"\nFile Error: {error}")

    except Exception as error:
        print(f"\nUnexpected Error: {error}")


# Run comparison
run_model_comparison()
```
